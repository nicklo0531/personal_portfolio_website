```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
skills = ['Python', 'Pandas', 'SQL', 'Matplotlib', 'NumPy', 'HTML', 'Machine Learning', 'Statistical Methods', 'RStudio', 'Tableau', 'GCP', 'Java', 'Javascript']
ratings = [4.6, 4.8, 4.8, 4.6, 4.6, 4.5, 4.5, 4.3, 3.8, 3.5, 3.5, 3.5, 3.2]
data = pd.DataFrame({'Skills': skills, 'Ratings': ratings})
plt.figure(figsize=(10,6))
sns.barplot(x='Ratings', y='Skills', data=data, palette='Blues_d')
plt.title('Skills and Ratings')
plt.xlabel('Rating (1-5)')
plt.ylabel('Skills')
plt.show()
```
